# House Price Prediction system

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

Importing the data

```python
data_url = "project_3_boston_data.txt"
house_price_df = pd.read_csv(data_url, sep="\s+", names=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'], header=None)
```

```python
house_price_df.head()
```

```python
house_price_df.shape
```

```python
house_price_df.isnull().sum()
```

```python
# statistical measures
house_price_df.describe()
```

Understanding the correlations between various features in the dataset

    1. Positive correlation

    2. Negative correlation 

```python
correlation = house_price_df.corr()
```

```python
# constructing heatmap to understand the correlation
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```

Splitting the data and target

```python
X = house_price_df.drop(['MEDV'], axis=1)
Y = house_price_df['MEDV']
```

```python
print(X)
print(Y)
```

Train test Spliting

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
from xgboost import XGBRegressor

# loading model
model = XGBRegressor()
```

```python
# training model
model.fit(X_train, Y_train)
```

Evaluation :

```python
# training data accuracy :
Y_train_predicted = model.predict(X_train)

# accuracy_train = metrics.accuracy_score(Y_train, Y_train_predicted)
rms_err_train = metrics.r2_score(Y_train, Y_train_predicted)
avg_arr_train = metrics.mean_absolute_error(Y_train, Y_train_predicted)

# print(accuracy_train)
print(rms_err_train)
print(avg_arr_train)
```

```python
plt.scatter(Y_train, Y_train_predicted)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

```python
# training data accuracy :
Y_test_predicted = model.predict(X_test)

# accuracy_train = metrics.accuracy_score(Y_train, Y_train_predicted)
rms_err_test = metrics.r2_score(Y_test, Y_test_predicted)
avg_arr_test = metrics.mean_absolute_error(Y_test, Y_test_predicted)

# print(accuracy_train)
print(rms_err_test)
print(avg_arr_test)
```

```python
plt.scatter(Y_test, Y_test_predicted)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
